```python
# Agora voce vai pegar o arquivo de Vendas-Dez.xlsx e juntar com a sua tabela normal(pd.concat) #

vendas_dez_df = pd.read_excel('Vendas-Dez.xlsx')

vendas_df = pd.concat([vendas_df, vendas_dez_df], ignore_index=True)
display(vendas_df)
```

```python
# depois de juntas dois arquivos, preencha os campos NaN com as comissoes e imposto 0 #

vendas_df['Comissão'] = vendas_df['Comissão'].fillna(vendas_df['Valor Final'] * 0.05)
vendas_df['Imposto'] = vendas_df['Imposto'].fillna(0)
display(vendas_df)
```

```python
# Exclua a coluna de Imposto #

vendas_df = vendas_df.drop('Imposto', axis=1)
# vendas_df = vendas_df.drop(columns=['Imposto'])
display(vendas_df)
```
